```python
#import library
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.nobroker.in/flats-for-sale-in-electronic_city_bangalore')
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
house_details=soup.find_all('h2',class_='heading-6 font-semi-bold nb__1AShY')
house_details
```

```python
House_details=[]
for i in house_details:
    House_details.append(i.text)
    
House_details
```

```python
emi=soup.find_all('div', class_='nb__2NPHR')
emi
```

```python
EMI=[]
for i in emi:
    EMI.append(i.text)
EMI
```

```python
area=soup.find_all('div',class_='nb__3oNyC')
area
```

```python
Area=[]
for i in area:
    Area.append(i.text)
Area
```

```python
location=soup.find_all('div',class_="nb__2CMjv")
location
```

```python
Location=[]
for i in location:
    Location.append(i.text)
Location
```

```python
price=soup.find_all('div',class_='nb__2NPHR')
price
```

```python
Price=[]  #here monthly EMI total_EMI ,bulit area are mixed so lets segregate each one
for i in price:
    
    Price.append(i.text)
    
print(Price)
```

```python
Price.remove('Title not verifiedUnder Loan₹ ')
print(Price)
```

```python
Price.remove('Title not verifiedNot under loan₹ ')
print(Price)
```

```python
built_area=[]
month_emi=[]
total_emi=[]
for i in Price:
    
    if 'sqftBuiltup'  in i:
        
        built_area.append(i)
    
    elif 'EMI' in i:
        
        month_emi.append(i)
    else :
        
        
        total_emi.append(i)
print(built_area,month_emi,total_emi)
```

```python
Total_EMI=[]
Title=[]
for i in total_emi:
    
    if 'Title' in i:
        Title.append(i)
    else:
        Total_EMI.append(i)
print(Total_EMI,Title)
```
