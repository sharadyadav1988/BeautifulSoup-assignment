```python
#importing the libraries
from bs4 import BeautifulSoup
import requests
```

```python
#get url response
page=requests.get("https://bookpage.com/")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
book=soup.find_all('h4',class_="italic")
book
```

```python
book_name=[]
for i in book:
    book_name.append(i.text.replace('\n',''))
book_name    
```

```python
author_name=soup.find_all('p',class_="sans bold")
author_name
```

```python
author=[]
for i in author_name:
    author.append(i.text.replace('\n',''))
author  
```

```python
genre=soup.find_all('p',class_='genre-links hidden-phone')
genre
```

```python
genre_list=[]
for i in genre:
    genre_list.append(i.text)
    
genre_list
```

```python
genre_list=[]
for i in genre:
    genre_list.append(i.text.replace('\n',''))
    
genre_list
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Book']=book_name
df['Author']=author
df['Gener']=genre_list
df
```
