```python
titles=soup.find_all('div' ,class_='heading_4_5 profile')
titles
```

```python
#now we have all tag in which there are job titles
#now we will extract the next form  these tags one by one looping over these tags 
job_titles=[]  #empty list to store
for i in titles:
    job_titles.append(i.text)
    
job_titles    
```

```python
#now the text of the job tag title is inside the tags extracted above 
job_titles=[] #empty list in this list the data will store
for i in titles:
    job_titles.append(i.text.replace('\n',''))
    
job_titles    
```

```python
#so lets extract all tags haveing compnay names
companies=soup.find_all('a',class_="link_display_like_text")
companies
```

```python
#now we all have tags which have company name
company_name=[]  #empty list to store all data
#now we extract text  from these tags one by one thru  for loop
for i in companies:
    company_name.append(i.text.strip())
    
company_name    
```

```python
len(company_name)
```

```python
#so lets extract all the tag ctc
ctc=soup.find_all('div', class_='other_detail_item')
ctc
```

```python
# CTC and apply date both are mixed lest segregate and make a list


ctc_list=[]
for i in ctc:
    ctc_list.append(i.text.replace('\n',''))
    
ctc_list  
```

```python
CTC_list=[]
Apply_list=[]
for i in ctc_list:
    if 'CTC' in i:
        CTC_list.append(i)
    elif 'Apply' in i:
        Apply_list.append(i)
print(CTC_list)
print(Apply_list)
```

```python
#now check all len ,must be same lenght to make dataframe otherwise mismatch

print(len(job_titles),len(company_name),len(CTC_list),len(Apply_list))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['title']=job_titles
jobs['company']=company_name
jobs['CTC']=CTC_list
jobs['Apply Date']=Apply_list
```
