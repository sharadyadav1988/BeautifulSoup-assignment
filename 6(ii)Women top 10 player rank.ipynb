```python
#importing the libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
no1_palyer=soup.find('div',class_='rankings-block__banner--name-large')
A=no1_palyer.text
A=[A]
A
```

```python
other_palyer=soup.find_all('td',class_='table-body__cell rankings-table__name name')
other_palyer
```

```python
player=[]
for i in other_palyer:
    player.append(i.text.replace('\n',''))
player
```

```python
B=player[0:9] #select top 9
B
```

```python
C=A+B #add no 1 with other top 9 to make to 10 palyer
C
```

```python
country=soup.find('div',class_='rankings-block__banner--nationality')
P=country.text.replace('\n','')
P=[P]
P
```

```python
other_country=soup.find_all('span',class_='table-body__logo-text')
other_country
```

```python
top_othercountry=[]
for i in other_country:
    top_othercountry.append(i.text)
    
top_othercountry
```

```python
D=top_othercountry[0:9]
D
```

```python
E=P+D
print(E)
```

```python
rating_top=soup.find('div',class_='rankings-block__banner--rating')
P=rating_top.text
P=[P]
P
```

```python
rating_other=soup.find_all('td',class_='table-body__cell rating')
rating_other
```

```python
other9_country=[]
for i in rating_other:
    other9_country.append(i.text)
F=other9_country
F=F[0:9]
print(F)
```

```python
G=P+F  #rating top 10 team after adding no 1 with other 9
G
```

```python
best=soup.find('div',class_='rankings-block__career-best')
O=best.text.replace('\n','')
O=[O]
O
```

```python
other_best=soup.find_all('td',class_='table-body__cell u-text-right u-hide-phablet')
other_best
```

```python
other_best9=[]
for i in other_best:
    other_best9.append(i.text.replace('\n',''))
    
H=other_best9
H=H[0:9]
H
```

```python
I=O+H #add top and other 9 best 
I
```
